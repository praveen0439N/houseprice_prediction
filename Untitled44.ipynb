```python
import matplotlib.pyplot as plt
plt.bar(df['bedrooms'],df['price'],color='red')
plt.xlabel("Bedrooms")
plt.ylabel("Count")
plt.title("Number of Houses by Bedrooms")
plt.show()
```

```python
plt.scatter(df['area'], df['price'],color='pink')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area vs Price')
```

```python
plt.hist(df['price'],bins=40,color='green',edgecolor='blue',alpha=0.7)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

```python
plt.boxplot(df['parking'])
plt.title('Price Boxplot')
plt.show()
```

```python
plt.barh(df['area'],df['stories'],color='violet')
plt.xlabel('Area')
plt.ylabel('Price')
```
